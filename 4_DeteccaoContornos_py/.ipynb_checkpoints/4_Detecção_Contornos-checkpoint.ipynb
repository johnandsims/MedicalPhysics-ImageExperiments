## Calculation of gradient and angle $\alpha$ with simple derivative kernel

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
import math
I = np.zeros([8,8],dtype='uint8')

wx = ( [[ -1 , 1 ]] ) #list object
wx = np.array([[-1,1]]) #np array
wy=wx.transpose() # ust be np array-  does not work with list object.
print('wx = ', wx)
print('wy = ', wy)
I[3:,3:]= 1
I[3,3:5]=0
I[4,3] = 0
print(I)
plt.imshow(I)
plt.colorbar()
plt.show()

gx = ndimage.correlate(I,wx, mode='reflect')
plt.imshow(gx)
plt.colorbar()
plt.title('gx')
plt.show()

gy = ndimage.correlate(I,wy, mode='reflect')
plt.imshow(gy)
plt.colorbar()
plt.title('gy')
plt.show()

print(f"gx ={gx}")
print(f" gy = {gy}")
alpha = 180/math.pi * np.arctan2(gy,gx)
print('angle, $alpha$ = ', alpha)
```

## Calculation of gradient and angle $\alpha$ with sobel kernel

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
import math
I = np.zeros([8,8],dtype='uint8')

I[3:,3:]= 1
I[3,3:5]=0
I[4,3] = 0
print(I)
plt.imshow(I)
plt.colorbar()
plt.show()

gx = ndimage.sobel(I, axis=1) #vertical
plt.imshow(gx)
plt.colorbar()
plt.title('gx')
plt.show()

gy = ndimage.sobel(I, axis=0) #horizontal
plt.imshow(gy)
plt.colorbar()
plt.title('gy')
plt.show()

magGrad = (gx**2 + gy**2)**0.5
print("magnitude of gradient: ",magGrad)

print(f"gx ={gx}")
print(f" gy = {gy}")
alpha = 180/math.pi * np.arctan2(gy,gx)
print('angle, alpha = ',alpha)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Default is to use 'xy' style coords (i.e. origin in bottom left)
#  To use matrix coords, we invert yaxis and use angles='xy' in quiver.

# 1.  Use Sobel (gx and gy calculated in last cells)
fig, ax = plt.subplots(figsize=(7,7))
ax.invert_yaxis()
q = ax.quiver(gx,gy, angles='xy')
plt.title('Gradient using Sobel kernel')
qk = ax.quiverkey(q, 0.8, 0.9, 2, '2 unit', labelpos='E',
                   coordinates='figure')
plt.show()

# 2. Use function gradient (supplied by python.)
v, u = np.gradient(I, .2, .2)
fig, ax = plt.subplots(figsize=(7,7))
ax.invert_yaxis()
q = ax.quiver(u,v, angles='xy')
qk = ax.quiverkey(q, 0.8, 0.9, 1, '1 unit', labelpos='E',
                   coordinates='figure')
plt.title('Gradient using Numpy function')
plt.show()

magGrad = (u**2 + v**2)**0.5
print("magnitude of gradient: ",magGrad)


```

## Quiver applied to Gaussian function


```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)

X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)

fig, ax = plt.subplots(figsize=(9,9))
ax.imshow(z,cmap='gray')


dy,dx = np.gradient(z)


fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(12,12))
#fig.suptitle('Gradients in x and y')

ax1.imshow(dx)
ax1.set_title('gradient in x, dx')

ax2.imshow(dy)
ax2.set_title('gradient in y, dy')



fig, ax = plt.subplots(figsize=(9,9))
ax.invert_yaxis()
q = ax.quiver(X,Y,dx,dy, angles='xy')
plt.title('Gradient using Sobel kernel')
qk = ax.quiverkey(q, 0.8, 0.9, 0.1, '0.1 unit', labelpos='E',
                   coordinates='figure')
plt.show()



ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

plt.show()
```

## Laplacian of the Gaussian border detection

Steps for LoG border detection:

    Apply LoG on the image. This can be done in two ways:
        First, apply Gaussian and then Laplacian or
        Convolve the image with LoG kernel directly
    Find the zero crossings in the image
    Threshold the zero crossings to extract only the strong edges.

```python
import cv2
import numpy as np
 
# Load the image 
from skimage import data
img=data.coins() 
    
print(img)
# Apply Gaussian Blur
blur = cv2.GaussianBlur(img,(3,3),0)
 
# Apply Laplacian operator in some higher datatype
laplacian = cv2.Laplacian(blur,cv2.CV_64F)
```

clip the negative values to find the zero crossings.

```python
# But this tends to localize the edge towards the brighter side.
laplacian1 = laplacian/laplacian.max()
 
cv2.imshow('a7',laplacian1)
cv2.waitKey(0)
```

```python
def Zero_crossing(image):
    z_c_image = np.zeros(image.shape)
    
    # For each pixel, count the number of positive
    # and negative pixels in the neighborhood
    
    for i in range(1, image.shape[0] - 1):
        for j in range(1, image.shape[1] - 1):
            negative_count = 0
            positive_count = 0
            neighbour = [image[i+1, j-1],image[i+1, j],image[i+1, j+1],image[i, j-1],image[i, j+1],image[i-1, j-1],image[i-1, j],image[i-1, j+1]]
            d = max(neighbour)
            e = min(neighbour)
            for h in neighbour:
                if h>0:
                    positive_count += 1
                elif h<0:
                    negative_count += 1
 
 
            # If both negative and positive values exist in 
            # the pixel neighborhood, then that pixel is a 
            # potential zero crossing
            
            z_c = ((negative_count > 0) and (positive_count > 0))
            
            # Change the pixel value with the maximum neighborhood
            # difference with the pixel
 
            if z_c:
                if image[i,j]>0:
                    z_c_image[i, j] = image[i,j] + np.abs(e)
                elif image[i,j]<0:
                    z_c_image[i, j] = np.abs(image[i,j]) + d
                
    # Normalize and change datatype to 'uint8' (optional)
    z_c_norm = z_c_image/z_c_image.max()*255
    z_c_image = np.uint8(z_c_norm)
 
    return z_c_image
```

```python
I = Zero_crossing(laplacian)
print(I)
plt.imshow(laplacian,cmap='gray')
plt.show()
```
