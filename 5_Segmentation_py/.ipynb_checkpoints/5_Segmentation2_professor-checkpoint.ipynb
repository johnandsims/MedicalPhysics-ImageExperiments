### Create Segmentation masks from manually segmented images

```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt


# Get Nifti kidney segmentation and show image.
# Files are in Images/renal on tidia.
im = nib.load('../../Images/renal/Kidney_Areas.nii')
data = im.get_data()
I = data[:,:,30]
plt.imshow(I)
plt.colorbar()
plt.show()
```

### Now get right and left kidneys as separate masks:

```python
RightKid = np.where(I==1,1,0)
LeftKid = np.where(I==2,1,0)

fig, ((ax1, ax2)) = plt.subplots(1,2,figsize=(15,15))
ax1.imshow(LeftKid,cmap='gray')
ax1.set_title(f'Left Kidney')
ax2.imshow(RightKid,cmap='gray')
ax2.set_title(f'Right Kidney')
```

### Read in kidney images.

```python


Kid = nib.load('../../Images/renal/Kidneys.nii')
row,col,slices = np.shape(Kid)
print(f'row={row}, col={col}, slices={slices}')

data = Kid.get_data()
```

###  Discard the first 30 images.

```python
Kim = data[:,:,30:]

# Get dimensions of kidney image sequence
row,col,slices = np.shape(Kim)
print(f'row={row}, col={col}, slices={slices}')
```

## Now blur in spatial dimension (x and y).

```python
import cv2
# Use Gaussian 3x3 filter, sigma = 1.2
KimBlurS = cv2.GaussianBlur(Kim,(3,3),0.7)

# Show images:
fig, ((ax1, ax2, ax3)) = plt.subplots(1,3,figsize=(15,15))
ax1.imshow(Kim[:,:,0],cmap='gray')
ax1.set_title(f'First image kidneys') 
ax2.imshow(KimBlurS[:,:,0],cmap='gray')
ax2.set_title(f'First image blurred')
ax3.imshow(KimBlurS[:,:,29],cmap='gray')
ax3.set_title(f'Last image blurred')

plt.show()
```

## Blur in temporal direction (along time with filter [1,2,1]/4)

```python
#Create new matrix to hold temporal data.  We lose a time sample at the beginning and end. (slices-2)
KimBlurT = np.zeros([row,col,slices-2])

for i in range(0,slices-2):
    KimBlurT[:,:,i] = KimBlurS[:,:,i]/4.0 + 2*KimBlurS[:,:,i+1]/4.0 + KimBlurS[:,:,i+2]+4.0
    
# Show images:
fig, ((ax1, ax2, ax3)) = plt.subplots(1,3,figsize=(15,15))
ax1.imshow(Kim[:,:,0],cmap='gray')
ax1.set_title(f'First image kidneys')
ax2.imshow(KimBlurT[:,:,0],cmap='gray')
ax2.set_title(f'First image blurred')
ax3.imshow(KimBlurT[:,:,27],cmap='gray')
ax3.set_title(f'Last image blurred')

plt.show()
```

## Now apply masks to image set.

```python
# We changed the size of the image matrix in KimBlurT
row,col,slices = np.shape(KimBlurT)

# Create empty arrays for right and left kidney images.
Lim = np.zeros([row,col,slices])
Rim = np.zeros([row,col,slices])

# Loop through all remaining slices.
for i in range(0,slices):
    Lim[:,:,i] = LeftKid * KimBlurT[:,:,i]
    Rim[:,:,i] = RightKid * KimBlurT[:,:,i]
    fig, ((ax1, ax2)) = plt.subplots(1,2,figsize=(9,9))
    ax1.imshow(Lim[:,:,i],cmap='gray')
    ax1.set_title(f'Left Kidney frame {i}')
    ax2.imshow(Rim[:,:,i],cmap='gray')
    ax2.set_title(f'Right Kidney frame {i}')

plt.show()
```

## Now sum the set of masked kidneys to get total count for each instant.

```python
# Get sum of count in each kidney over the slices (producing a 1d vector)
# Sum kidney image over axis 1, then the result over 0.
Lsum=Lim.sum(axis=1)
Lsum=Lsum.sum(axis=0)

Rsum=Rim.sum(axis=1)
Rsum=Rsum.sum(axis=0)

# Plot the results (Right and left kidney sum counts on same figure)
plt.plot(Lsum, label='Left Kidney')
plt.plot(Rsum, label='Right Kidney')
plt.legend(loc="upper right")
plt.xlabel('Frames (time)')
plt.ylabel('Count (sum)')
plt.title('Sum of radioactive counts in left and right kidneys over time')
plt.show()
```

## Difference image:

```python
diff = Lim[:,:,3]-Lim[:,:,4]
plt.imshow(diff)
plt.show()

row,col,slices = np.shape(Lim)

Im3 = np.zeros([row,col])
Im3[diff>0.015] = 1

plt.imshow(Im3)
plt.show()
```
