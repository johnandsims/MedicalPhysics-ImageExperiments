```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 10 16:23:28 2020

@author: john
"""
import numpy as np
import matplotlib.pyplot as plt

G = 400
R1 = 80.0

x = np.arange(-G, G)
y = np.arange(-G, G)
X,Y = np.meshgrid(x,y)
r,c = np.shape(X)
I = np.zeros([r,c])
I[X**2.0 +Y**2.0 < R1**2.0] = 1

plt.imshow(I)
plt.colorbar()
plt.show()

```

```python
# Adapted from https://scikit-image.org/docs/dev/auto_examples/transform/plot_radon_transform.html

from skimage.io import imread
from skimage.transform import radon, iradon

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Original")
ax1.imshow(I, cmap=plt.cm.Greys_r)

theta = np.linspace(0., 180., max(I.shape), endpoint=False)
sinogram = radon(I, theta=theta, circle=True)
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')

fig.tight_layout()
plt.show()

```

```python
a,b=np.shape(sinogram)
print(a,b)
plt.plot(sinogram[:,0])
plt.title('Sinogram at 0$^o$')
plt.show()
plt.plot(sinogram[:,400])
plt.title('Sinogram at 90$^o$')
plt.show()
```

## Create new image, to increase complexity.

```python
import numpy as np
import matplotlib.pyplot as plt

G = 400
R1 = 80.0
R2 = 30.0;
Xoffset = -175.0;

x = np.arange(-G, G)
y = np.arange(-G, G)
X,Y = np.meshgrid(x,y)
r,c = np.shape(X)
I = np.zeros([r,c])
I[X**2.0 +Y**2.0 < R1**2.0] = 1
I[(X - Xoffset) **2.0 + Y**2.0 < R2**2.0] = 1;

plt.imshow(I)
plt.colorbar()
plt.show()
```

```python
from skimage.io import imread
from skimage.transform import radon, iradon

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Original")
ax1.imshow(I, cmap=plt.cm.Greys_r)

theta = np.linspace(0., 180., max(I.shape), endpoint=False)
sinogram = radon(I, theta=theta, circle=True)
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')

fig.tight_layout()
plt.show()
```

```python
a,b=np.shape(sinogram)
print(a,b)
plt.plot(sinogram[:,0])
plt.title('Sinogram at 0$^o$')
plt.show()
plt.plot(sinogram[:,400])
plt.title('Sinogram at 90$^o$')
plt.show()
```

```python
a = sinogram[:,0]
b = sinogram[:,400]

A = np.array([a]*800)
B = np.array([b]*800)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Radon transform\n(Sinogram) (0$^o$)")
ax1.set_xlabel("Projection angle (deg)")
ax1.set_ylabel("Projection position (pixels)")
ax1.imshow(A)

ax2.set_title("Radon transform\n(Sinogram) (90$^o$)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(B)

fig.tight_layout()
plt.show()
```

## Shepp Logan phantom:

```python
from skimage.io import imread
from skimage.data import shepp_logan_phantom
from skimage.transform import radon, rescale

image = shepp_logan_phantom()
image = rescale(image, scale=0.4, mode='reflect', multichannel=False)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Original")
ax1.imshow(image, cmap=plt.cm.Greys_r)

theta = np.linspace(0., 180., max(image.shape), endpoint=False)
sinogram = radon(image, theta=theta, circle=True)
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')

fig.tight_layout()
plt.show()
```
